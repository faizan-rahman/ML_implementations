```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
irdata = pd.read_csv('iris_data.txt', sep=",", header=None)
irdata.columns = ["sepal length in cm", "sepal width in cm", "petal length in cm", "petal width in cm", "class"]
```

```python
irdata.head()
```

```python
irdata.describe()
```

```python
sns.pairplot(data = irdata, hue = "class", palette = 'colorblind')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = irdata[['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm']]
y = irdata['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
